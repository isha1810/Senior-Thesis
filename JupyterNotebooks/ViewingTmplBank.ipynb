```python
from glue.ligolw import ligolw
from glue.ligolw import table
from glue.ligolw import lsctables
from glue.ligolw import ilwd
from glue.ligolw import utils as ligolw_utils
import pycbc.strain
import pycbc.psd
import numpy as np
import matplotlib.pyplot as plt
```

#### For Mass Range of 2-10

```python
bank_filename = 'GreaterRangeNonSpin.xml'
xml_doc = ligolw_utils.load_filename(bank_filename, contenthandler=table.use_in(ligolw.LIGOLWContentHandler), verbose=True)
sngl_inspiral_table = lsctables.SnglInspiralTable.get_table(xml_doc)
```

```python
m1 = np.array([sngl_wf.mass1 for sngl_wf in sngl_inspiral_table])
m2 = np.array([sngl_wf.mass2 for sngl_wf in sngl_inspiral_table])
tau0 = np.array([sngl_wf.tau0 for sngl_wf in sngl_inspiral_table])
tau2 = np.array([sngl_wf.tau2 for sngl_wf in sngl_inspiral_table]) #all 0s
tau3 = np.array([sngl_wf.tau3 for sngl_wf in sngl_inspiral_table])
mchirp = np.array([sngl_wf.mchirp for sngl_wf in sngl_inspiral_table])
eta = np.array([sngl_wf.eta for sngl_wf in sngl_inspiral_table])
mtotal = np.array([sngl_wf.mtotal for sngl_wf in sngl_inspiral_table])
q = m1/m2
```

```python
print("Total {} test-points".format(len(sngl_inspiral_table)))
```

##### Plotting in different coords

```python
plt.figure()
plt.plot(m1, m2, 'kx', alpha=0.3)
plt.grid()
plt.title("Template Bank")
plt.xlabel('mass 1')
plt.ylabel('mass 2')
```

```python
plt.figure()
plt.plot(tau0, tau3, 'kx', alpha=0.3)
plt.grid()
plt.title("Template Bank")
plt.xlabel('tau0')
plt.ylabel('tau3')
```

```python
plt.figure()
plt.plot(mchirp, eta, 'kx', alpha=0.3)
plt.grid()
plt.title("Template Bank")
plt.xlabel('mchirp')
plt.ylabel('eta')
```

```python
plt.figure()
plt.plot(mtotal, q, 'kx', alpha=0.3)
plt.grid()
plt.title("Template Bank")
plt.xlabel('mtotal')
plt.ylabel('q')
```
